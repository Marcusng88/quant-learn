## coding challenge

```python
symbol='GE'
start='2015-01-02'
end= '2020-12-31'

df = yf.download(symbol,start,end,multi_level_index = False)
```

```python
df
```

```python
close = df.Close
```

```python
norm = close.div(close.iloc[0])
# normalized price on 2020-12-30 is 0.509
norm.loc['2020-12-30']
```

```python
norm.column = ['Close Price']
norm
```

```python
fig, ax = plt.subplots()
sns.set_style('darkgrid')
ax.set_xlabel('time')
ax.set_ylabel('Normalized price')
ax.set_title('Returns of GE')
ax.plot(norm)
plt.show()
```
